```python
for i in range(0,572):
    if(data['Location'][i] == "Other"):
        data['Location'][i] = "East Europe"
        
    if(data['Location'][i] == "Singapore" or data['Location'][i] == "United Arab Emirates" or data['Location'][i] == "Pakistan"):
        data['Location'][i] = "Asia"
        
    if(data['Location'][i] == "United States"):
        data['Location'][i] = "North America"
        
    if(data['Location'][i] == "United Kingdom"):
        data['Location'][i] = "West Europe"
    
    
data.head()
```
