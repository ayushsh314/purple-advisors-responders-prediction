```python
import pandas as pd
url = f'https://docs.google.com/spreadsheets/d/1vKAIohc1IYOc6E3Y2YLRzAgLcwswKrSKLXX2J47rhjE/gviz/tq?tqx=out:csv&sheet=Updated_Data_Sheet'
df = pd.read_csv('temp_service.csv')
```

```python
length = len(df['Industry Classification'])
```

```python
length
```

```python
for i in range(0,length):
    if((df['Industry Classification'][i] == 'Product') & (df['Number of employees'][i] < 40)):
        df['compute_rev'][i] = df['Number of employees'][i] * 15000
        
    elif((df['Industry Classification'][i] == 'Product') & (df['Number of employees'][i] in range(40,101))):
        df['compute_rev'][i] = (df['Number of employees'][i])/2 * 1000 * df['Number of employees'][i]
    
    elif((df['Industry Classification'][i] == 'Product') & (df['Number of employees'][i] > 100)):
        df['compute_rev'][i] = (df['Number of employees'][i]) * 50000
        
        # f(x) = ((b-a)(x - min))(max-min) + a
        
        # [min, max] --> [a,b]
        # x=67
        
        # x = 59(arr/emp) > x=88(arr/emp)
        
        # x=64
        # [50,100] --> [25-50k]
        # (25*14)/50 + 25   * 1000 * no. of employees = ARR
```
